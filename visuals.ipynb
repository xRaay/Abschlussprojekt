```python
import pandas as pd

df = pd.read_csv(data)

df = df.drop('Unnamed: 0', axis=1)

df.info()

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#set style of plots
sns.set_style('whitegrid')

#plot gender distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df, palette='coolwarm')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

#Age Group Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Age Group', data=df, palette='viridis', order = df['Age Group'].value_counts().index)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

#Country of Birth Distribution
plt.figure(figsize=(10, 6))
df_country = df['Country Of Birth'].value_counts().head(10)
sns.barplot(x=df_country.values, y=df_country.index, palette='magma')
plt.title('Top 10 Countries of Birth')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Tenure Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Tenure'], bins=30, kde=True, color='darkred')
plt.title('Tenure Distribution')
plt.xlabel('Years')
plt.ylabel('Density')
plt.show()

# Age at Start Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age at start'], bins=30, kde=True, color='green')
plt.title('Age at Start Distribution')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np 

plt.figure(figsize=(12, 8))
sns.barplot(x='rating_num', y='division', data=df, errorbar=None, palette='Spectral', estimator=np.mean)
plt.title('Performance Ratings by Division')
plt.xlabel('Average Performance Rating')
plt.ylabel('Division')
plt.show()
```
